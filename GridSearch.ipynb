```python
import joblib
import pandas as pd
import numpy as np
from time import time

x_train=pd.read_csv("drive/Shareddrives/major_project/data/x_train.csv")
y_train=pd.read_csv("drive/Shareddrives/major_project/data/y_train.csv")
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

```python
#Grid Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 130, num = 12)]
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
def print_res(results):
  print("Best Params: {}\n".format(results.best_params_))
  means=results.cv_results_['mean_test_score']
  stds=results.cv_results_['std_test_score']
  for mean,std,params in zip(means,stds,results.cv_results_['params']):
    print('{} (+/-{}) for {}'.format(round(mean,3),round(std*2,3),params))
```

```python
# Grid search of parameters
from sklearn.model_selection import GridSearchCV
rf_grid = GridSearchCV(estimator = rf, param_grid = random_grid)
start=time()
rf_grid.fit(x_train,y_train.values.ravel())
end=time()
sec=(end-start)%60
min=(end-start)//60
print_res(rf_grid)
print("Time Elapsed : {}mins {}s".format(round(min,3),round(sec,3)))
```

```python
rfg_model=rf_grid.best_estimator_
print(rfg_model)
```

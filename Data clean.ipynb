```python
#delete column 'torque,name'
#create dataframe and delete column 'torque,name'
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("drive/Shareddrives/major_project/car.csv")
df.drop(['torque','name'],inplace=True,axis=1)
df.shape
```

```python
# remove rows with empty columns
df.dropna(inplace=True)
df.shape
```

```python
#split mileage to give only the value 
new=df['mileage'].str.split(" ",n=1,expand=True)
df['mileage']=new
#split engine to give only the value 
new=df['engine'].str.split(" ",n=1,expand=True)
df['engine']=new
#split max_power to give only the value 
new=df['max_power'].str.split(" ",n=1,expand=True)
df['max_power']=new
```

```python
#Adding current year column 
df['Current Year']=2021
# Find no. of years
df['no. of years']=df['Current Year']-df['year']
# drop original and current year
df.drop(['year'],axis=1,inplace=True)
df.drop(['Current Year'],axis=1,inplace=True)
df.head()
```

```python
#convert objects to numerical values
df['mileage']=df['mileage'].astype(float)
df['seats']=df['seats'].astype(int)
df['engine']=df['engine'].astype(float)
df['max_power']= pd.to_numeric(df['max_power'],errors='coerce')
df.dropna(inplace=True)
```

```python
#one hot encoding
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
#independent and dependent features
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=42)
x_test,x_val,y_test,y_val=train_test_split(x_test,y_test,test_size=0.4,random_state=42)
```
